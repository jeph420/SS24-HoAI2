        <li>Define the following two $3 \times 3$ filters (shown in the formulae below) and apply them on 12 random images $A$ from the above data set (with $*$ as the convolution and $\sigma{}$ as the sigmoid operation) to produce the following 4 outputs $G_x, G'_x, G_y, G'_y$:</li>
    </ul>
    <p>
        \begin{equation}G_x = \left(
            \begin{array}{rrr}
                -5 & 0 & 5 \\
                -5 & 0 & 5 \\
                -5 & 0 & 5
            \end{array}\right) * A
            \qquad
            G'_x = \sigma (G_x)
        \end{equation}
    </p>
    <p>
        \begin{equation}G_y = \left(
            \begin{array}{rrr}
                -5 & -5 & -5 \\
                 0 &  0 &  0 \\
                 5 &  5 &  5
            \end{array}\right) * A
            \qquad
            G'_y = \sigma (G_y)
        \end{equation}
    </p>
    <ul>
        <li>Hint: Make sure to exclude the class label column <i>digit</i> before processing your data.</li>
    </ul>
</div>

```python
# Your code
filter_x = np.asarray([
    [-5, 0, 5],
    [-5, 0, 5],
    [-5, 0, 5]
])
filter_y = np.asarray([
    [-5, -5, -5],
    [ 0,  0,  0],
    [ 5,  5,  5]
])
def sig(x):
    return 1/(1+np.exp(-x))

G_x = filter_x
G_y = filter_y
sig_G_x = sig(G_x)
sig_G_y = sig(G_y)
```

        <tr>

            <td>ReLU</td>
            <td>-</td>
        </tr>
        <tr>
            <td>3</td>
            <td>max pooling</td>
            <td>kernel size of $2\times{}2$</td>
        </tr>
        <tr>
            <td>4</td>
            <td>fully connected</td>

        </tr>
    </table>

```python
# Your code
class CNN(torch.nn.Module):
    
     def __init__(self):
         super().__init__()
         self.conv_stack = torch.nn.Sequential(
             torch.nn.Conv2d(1, 28, 3),  # Size changes from (1, 28, 28) to (10, 26, 26).
             torch.nn.ReLU(),
             torch.nn.MaxPool2d(2)  # Size changes from (10, 26, 26) to (10, 13, 13).
         )
         self.fc_stack = torch.nn.Sequential(
             torch.nn.Flatten(),  # Size changes from (10, 13, 13) to (10 * 13 * 13).
             torch.nn.Linear(28 * 13 * 13, 512),
             torch.nn.ReLU(),
             torch.nn.Linear(512, 10)
         )

     def forward(self, x: torch.Tensor) -> torch.Tensor:
         x = self.conv_stack(x)
         x = self.fc_stack(x)
         return x
```
